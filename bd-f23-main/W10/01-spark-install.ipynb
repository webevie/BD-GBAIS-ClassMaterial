# Installing PySpark

In this last section of the course we will use python to work with spark. To get started you need to have Java installed, environment variables added, and the pyspark and findspark packages installed. Follow the provided steps closely.

> NOTE: Before starting, make sure that Java isn't already installed. If it is, see the end of this document for instuctions on how to purge any java installation.

**Install the detault Java version for ubuntu (jre-1.11, often referred to as Java 11)**
```bash
sudo apt-get update
sudo apt-get upgrade
sudo apt-get 
sudo apt install default-jre
```

**Set Environmnet Variables**

Find where Java is installed 

```bash
which java
```

This should be a folder like the following...

```bash
/usr/lib/jvm/java-11-openjdk-amd64
```

Update the file /etc/environment

```bash
sudo nano /etc/environment
```

Add the line (note: the path should match the output from which java)

```bash
JAVA_HOME="/usr/lib/jvm/java-11-openjdk-amd64"
```

**Install PySpark**

```bash
conda install -c conda-forge pyspark
```

**Install sparkmagic**

```bash
pip install sparksql-magic
```

### Purging Java

> [WARNING!]   
> Only do this if before you begin the instructions above you have found an existing installation of Java

```bash
sudo apt-get purge java*
```

Check the following files, and remove any JAVA_HOME variable, or any mention of Java in teh $PATH variable.
